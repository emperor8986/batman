```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.read_csv('reco_test_hybrid_new5050.csv')
```

```python
ratings
```

```python
# pivot ratings into movie features
sku_features = ratings.pivot(
    index='SKU',
    columns='CID',
    values='RECO_METRIC'
).fillna(0)
```

```python
sku_features
```

```python
sku_similarity = cosine_similarity(sku_features)
```

```python
sku_similarity
```

```python
X = sku_features.T
```

```python
X
```

```python
sku_similarity_df = pd.DataFrame(sku_similarity, index = X.columns, columns = X.columns)
```

```python
sku_similarity_df
```

```python
def get_similar_skus(sku_name):
    similar_sku = sku_similarity_df[sku_name]
    similar_sku = similar_sku.sort_values(ascending = False)
    return similar_sku
```
